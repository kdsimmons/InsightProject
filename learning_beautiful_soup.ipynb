```python
trs = soup.findAll('tr')
```

```python
f = csv.writer(open("43rdCongress.csv", "w"))
f.writerow(["Name", "Years", "Position", "Party", "State", "Congress", "Link"])

for tr in trs:
    for link in tr.findAll('a'):
        fulllink = link.get('href')
    
    tds = tr.findAll('td')
    
    try:
        name = str(tds[0].getText().strip())
        years = str(tds[1].getText())
        position = str(tds[2].getText())
        party = str(tds[3].getText())
        state = str(tds[4].getText())
        congress = str(tds[5].getText())
        
        f.writerow([name, years, position, party, state, congress])
    except:
        print "bad tr string:", tds
        continue
```

```python
# tutorial 2: http://web.stanford.edu/~zlotnick/TextAsData/Web_Scraping_with_Beautiful_Soup.html
import urllib
r = urllib.urlopen('http://www.aflcio.org/Legislation-and-Politics/Legislative-Alerts').read()
soup = BeautifulSoup(r)
print type(soup)
```

```python
letters = soup.findAll('div', class_='ec_statements')
```

```python
len(letters)
```

```python
type(letters)
```

```python
letters[0]
```

```python
lobbying = {}
for el in letters:
    lobbying[el.a.getText()] = {}
```

```python
prefix = "www.aflcio.org"
for el in letters:
    lobbying[el.a.getText()]['link'] = prefix + el.a['href']
    date = el.find(id='legalert_date').getText()
    lobbying[el.a.getText()]['date'] = date
    
```

```python
for item in lobbying.keys():
    print(item + ': ' + '\n\t' + 'link: ' + lobbying[item]['link'] 
          + '\n\t' + 'date: ' + lobbying[item]['date'] + '\n\n')
```

```python
# alternate version using commands I've learned
letters_alt = soup.select('div.ec_statements')

alt_lobbying = []

for el in letters_alt:
    let = {}
    let['name'] = el.a.getText()
    let['link'] = prefix + el.a.attrs.get('href')
    let['date'] = el.select('div#legalert_date')[0].getText()
    
    alt_lobbying.append(let)
```
