```python
from sklearn.cluster import KMeans
import pymysql as mdb
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
mysqlauth = pd.DataFrame.from_csv('/home/kristy/Documents/auth_codes/mysql_user.csv')
sqluser = mysqlauth.username[0]
sqlpass = mysqlauth.password[0]

con = mdb.connect(user=sqluser, host="localhost", db="charity_data", password=sqlpass, charset='utf8')
```

```python
combined_panda = pd.DataFrame()
disease_list = ['alzheimer', 'blindness', 'breast_cancer', 'colon_cancer', 'crohn', 'diabetes',
                'dyslexia', 'leukemia', 'lung_cancer', 'multiple_sclerosis', 'osteoporosis',
                'parkinson']
disease_list = ['cancer']
with con:
    for disease in disease_list:
        pandadf = pd.read_sql("SELECT * FROM " + str(disease), con)
        combined_panda = pd.concat([combined_panda, pandadf], 0)
```

```python
X1 = combined_panda['twitter_followers'][:,np.newaxis]/500
X2 = combined_panda['cn_overall'][:,np.newaxis]
X = np.concatenate((X1, X2), axis=1)
```

```python
max_clust = 10
randcol = np.random.rand(max_clust,3)

for nclust in range(1,max_clust+1):
    k_means = KMeans(n_clusters=nclust)
    k_means.fit(X)
    y_pred = k_means.predict(X)
        
    plt.figure(nclust);
    for i in range(nclust):
        colnum = float(i)/float(nclust)
#        plt.plot(X1[y_pred==i], X2[y_pred==i], 'o', color=[colnum, 1.-colnum, randcol[i-1]], ms=10);
        plt.plot(X1[y_pred==i], X2[y_pred==i], 'o', color=randcol[i,:], ms=10);
    plt.axis([-5, 105,-5, 105]);
    plt.ylabel("Charity Navigator Score", fontsize=20);
    plt.xlabel("Twitter Followers (x 500)", fontsize=20);
    if nclust == 1:
        plt.title(str(nclust) + " cluster", fontsize=15)
    else:
        plt.title(str(nclust) + " clusters", fontsize=15)
```

```python
plt.figure(1);
plt.hist(combined_panda[combined_panda['twitter_followers']>0]['twitter_followers'][:,np.newaxis], bins=20);
plt.xlabel('Twitter Followers*', fontsize=20);
plt.ylabel('# of Charities', fontsize=20);
plt.title('Total: ' + str(sum(combined_panda['twitter_followers']>0)) + ' Charities', fontsize=15)

plt.figure(2);
plt.hist(combined_panda[combined_panda['cn_overall']>0]['cn_overall'][:,np.newaxis], bins=20);
plt.xlabel('Charity Navigator Score*', fontsize=20);
plt.ylabel('# of Charities', fontsize=20);
plt.title('Total: ' + str(sum(combined_panda['cn_overall']>0)) + ' Charities', fontsize=15)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
X_reduced = pca.transform(X)
print("Reduced dataset shape:", X_reduced.shape)
```

```python
plt.figure(figsize=[12,6])
plt.subplot(121)
plt.plot(X_reduced[:,0], X_reduced[:,1], 'o', color=[.23,.73,.2], ms=8);
plt.axis([-100, 50,-100, 50]);
plt.ylabel("PCA 1", fontsize=20);
plt.xlabel("PCA 2", fontsize=20);

plt.plot(plt.xlim(), plt.ylim(), 'k--');

plt.subplot(122)
plt.plot(X[:,0], X[:,1], 'o', ms=8);
plt.axis([-5, 105,-5, 105]);
plt.ylabel("original 1", fontsize=20);
plt.xlabel("original 2", fontsize=20);

plt.plot(plt.xlim(), plt.ylim(), 'k--');
```
