```python
# Set up connection to SQL
mysqlauth = pd.DataFrame.from_csv('/home/kristy/Documents/auth_codes/mysql_user.csv')
sqluser = mysqlauth.username[0]
sqlpass = mysqlauth.password[0]

con = mdb.connect(user=sqluser, host="localhost", db="charity_data", password=sqlpass, charset='utf8')
```

```python
# Import data from SQL
combined_panda = pd.DataFrame()
disease_list = ['alzheimer', 'blindness', 'breast_cancer', 'colon_cancer', 'crohn', 'diabetes',
                'dyslexia', 'leukemia', 'lung_cancer', 'multiple_sclerosis',
                'parkinson', 'prostate_cancer', 'cancer', 'tumor', 'melanoma', 'lymphoma']
with con:
    for disease in disease_list:
        pandadf = pd.read_sql("SELECT * FROM " + str(disease), con)
        combined_panda = pd.concat([combined_panda, pandadf], 0)
```

```python
# Put features into a numpy array
X = combined_panda[feature_names].values

# Get all features onto a fixed scale
X_scaled = preprocessing.scale(X)
```

```python
# Use PCA to reduce number of components
# Need to figure out how many components to get
pca = PCA(n_components=3)
pca.fit(X_scaled)
X_reduced = pca.transform(X_scaled)
print("Reduced dataset shape:", X_reduced.shape)

plt.figure(figsize=[18,6])
plt.subplot(131)
plt.plot(X_reduced[:,0], X_reduced[:,1], 'o')
plt.subplot(132)
plt.plot(X_reduced[:,0], X_reduced[:,2], 'o')
plt.subplot(133)
plt.plot(X_reduced[:,1], X_reduced[:,2], 'o')
```

```python
# Do k-means clustering on PCA features
max_clust = 10
randcol = np.random.rand(max_clust,3)

for nclust in range(1,max_clust+1):
    k_means = KMeans(n_clusters=nclust)
    k_means.fit(X_reduced)
    y_pred = k_means.predict(X_reduced)
        
    plt.figure(nclust, figsize=[6,6]);
    for i in range(nclust):
        colnum = float(i)/float(nclust)

        #plt.subplot(121)
        plt.plot(X_reduced[:,0][y_pred==i], X_reduced[:,1][y_pred==i], 'o', color=randcol[i,:], ms=8);
        #plt.axis([-100, 50,-100, 50]);
        plt.xlabel("PCA 1", fontsize=20);
        plt.ylabel("PCA 2", fontsize=20);        
        
        if nclust == 1:
            plt.title(str(nclust) + " cluster", fontsize=15)
        else:
            plt.title(str(nclust) + " clusters", fontsize=15)

```
